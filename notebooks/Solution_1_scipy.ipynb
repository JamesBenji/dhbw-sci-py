```python
#Numpy Pandas

import numpy as np
import pandas as pd
```

```python
# Exercise 1
elevations = np.array([23,24,589,90])

mean = np.mean(elevations)
std_deviation = np.std(elevations)

normalized_elevations = (elevations - mean) / std_deviation

new_mean = np.mean(normalized_elevations)
new_std_deviation = np.std(normalized_elevations)

print(f"The original array was {elevations}")
print(f"The normalized array is {normalized_elevations}")
print(f"The mean of the normalized array is {new_mean} versus that of the initial array {mean}")
print(f"The standard deviation of the normalized array is {new_mean} versus that of the initial array {std_deviation}")
```

```python
# Exercise 2
csv_url = "https://github.com/opengeos/datasets/releases/download/world/world_cities.csv"
cities = pd.read_csv(csv_url)

northern_hemisphere_cities = cities[cities['latitude'] > 0]

mean_latitude = northern_hemisphere_cities['latitude'].mean()

print(f"Average latitude of cities in the nothern hemisphere is {mean_latitude}")
```

```python
# Exercise 3
subset_cities = cities.head(4).copy()

def haversine_np(lat1, lon1, lat2, lon2):
    R = 6371.0
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = (np.sin(dlat / 2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2)**2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

lats = subset_cities['latitude'].values
lons = subset_cities['longitude'].values

lat1, lat2 = np.meshgrid(lats, lats)
lon1, lon2 = np.meshgrid(lons, lons)

distances = haversine_np(lat1, lon1, lat2, lon2)

dist_df = pd.DataFrame(
    distances,
    index=subset_cities['name'],
    columns=subset_cities['name']
)
print("Pairwise Distance Matrix (km):")
print(" ")
print(dist_df)
```
